```python
#https://github.com/opencv/opencv/tree/master/data/haarcascades
#얼굴 검출프로그램 만들기 opencv를 설치하면 얼굴 검출에 사용할 수 있는 캐스케이드 파일이 함께 OpenCV디렉터리에 저장됩니다.
#레파지토리를 보면 얼굴 검출 전용 캐스케이드 파일외 정면얼굴, 웃는 얼굴, 눈 , 몸 전체를 검출 하는 다양한 캐스케이드 파일이 있음
#haarcascade_frontalface_alt.xml -정면얼굴을 검출하는 xml사용 
#1.캐스캐이드 파일을 지정해서 검출기 만들기 2. 대상 이미지를 읽어 들이고 그레이스케일로 변환하기 3.얼굴 검출 실행하기
# 얼굴 검출을 하고 발견한 부분에 테두리를 두르는 프로그램을 만들어 보기

import matplotlib.pyplot as plt
import cv2

#캐스케이드 파일 지정해서 검출기 생성하기
cascade_file = "haarcascade_frontalface_alt.xml"
cascade = cv2.CascadeClassifier(cascade_file) #cv2.CascadeClassifier()를 사용하면 다양한 물체를 검출하기 위한 검출기를 만들 수 있음

#이미지를 읽어들이고 그레이스케일로 변환하기
img = cv2.imread("test2.jpg")  #이미지읽기
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #그레이스케일로 변환함 변환하는 이유는 식별기가 이미지 내부의 명암을 기반으로 얼굴인식

#얼굴 인식하기
face_list = cascade.detectMultiScale(img_gray, minSize=(150,150)) #얼굴검출을 실행 CascadeClassifier.detectMultiScale()메서드를 사용
#첫번째 매개변수로 그레이스케일 이미지데이터 전달 두번째 매개변수로 키워드 매개변수를 사용해 
#minSize를 지정함 이는 얼굴 인식 영역의 최소크기지정


#결과 확인하기  검출결과를 확인하고 리스트가 비어 있다면 메시지를 출력하고 종료함 
if len(face_list) == 0:
    print("실패")
    quit()
    
#인식한 부분만 표시  검출한 얼굴영역에 붉은색 테두리를 렌더링함 테두리렌더링은 cv2.rectangle()를 사용 이후 렌더링한 이미지를 파일로 출력하고 
#추가로 주피터 노트북에 출력함 

for (x,y,w,h) in face_list:
    print("얼굴의 좌표 =" , x, y ,w ,h)
    red = (0,0,255)
    cv2.rectangle(img,(x,y,), (x+w, y+h), red , thickness=20)
#이미지출력하기
cv2.imwrite("face-detect.png",img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#OPENCV로 모자이크 처리하기
#OpenCV에는 흐림 효과 또는 테두리 검출과 같은 필터함수가 있지만 모자이크 처리를 하는 함수는 없음 모자이크 처리하는
#함수를 mosaic()이라는 이름으로 정의하고 mosaic.py라는 모듈파일로 만들기

import cv2

def mosaic(img,rect,size):
    #모자이크 적용할 부분 추출하기
    (x1,y1,x2,y2) = rect
    w = x2 -x1
    h = y2 -y1
    i_rect = img[y1:y2, x1:x2]
    #축소하고 확대하기
    i_small = cv2.resize(i_rect,(size,size))
    i_mos = cv2.resize(i_small, (w,h), interpolation=cv2.INTER_AREA)
    
    #모자이크 적용
    img2 = img.copy()
    img2[y1:y2, x1:x2] = i_mos
    return img2



```

```python

import matplotlib.pyplot as plt
import cv2
from mosaic import mosaic as mosaic



#이미지를 읽어 들이고 모자이크 처리하기
img = cv2.imread("cat.jpg")
mos = mosaic(img,(50,50,450,450),10 )

#이미지 출력하기
cv2.imwrite("cat-mosaic.png", mos)
plt.imshow(cv2.cvtColor(mos,cv2.COLOR_BGR2RGB))
plt.show()
```
